```python
import regex as re
```

```python
with open("../loss_logs/custom_ds.txt", "r") as f:
    custom_ds_loss = re.findall(r"train_loss:\s*([\d.]+)", f.read())

custom_ds_loss = [float(s) for s in custom_ds_loss]
```

```python
with open("../loss_logs/ts_eng.txt", "r") as f:
    ts_eng_loss = re.findall(r"train_loss:\s*([\d.]+)", f.read())

ts_eng_loss = [float(s) for s in ts_eng_loss]
```

```python
with open("../loss_logs/ts_tr.txt", "r") as f:
    ts_tr_loss = re.findall(r"train_loss:\s*([\d.]+)", f.read())

ts_tr_loss = [float(s) for s in ts_tr_loss]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure()
plt.plot(custom_ds_loss, label="Custom TR dataset")
plt.plot(ts_tr_loss, label="TinyStories TR")
plt.plot(ts_eng_loss, label="TinyStories ENG")
plt.xlabel("Global Steps (divided by 200)")
plt.ylabel("Loss")
plt.ylim(bottom=0)
plt.title("Training Loss")
plt.legend()
plt.grid(True)
plt.show()
```

```python
custom2eng= [custom_ds_loss[i]/ts_eng_loss[i] for i in range(len(custom_ds_loss))]
tr2eng= [ts_tr_loss[i]/ts_eng_loss[i] for i in range(len(ts_tr_loss))]
eng2eng= [1.0]*len(ts_eng_loss)
```

```python
plt.figure()
plt.plot(custom2eng, label="Custom TR dataset")
plt.plot(tr2eng, label="TinyStories TR")
plt.plot(eng2eng, label="TinyStories ENG")
plt.xlabel("Global Steps (divided by 200)")
plt.ylabel("Loss ratio TinyStories ENG loss")
plt.ylim(bottom=0)
plt.legend()
plt.grid(True)
plt.show()
```
