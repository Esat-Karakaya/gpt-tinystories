```python
import matplotlib.pyplot as plt
```

```python
import re
```

```python
def plot(str):
    train_losses = [float(val) for val in re.findall(r"train_loss:\s*([\d\.]+)", str)]
    val_losses = [float(val) for val in re.findall(r"val_loss:\s*([\d\.]+)", str)]
    print(len(val_losses))
    print(val_losses)
    steps = [i*200 for i in range(len(val_losses))]

    plt.plot(steps, train_losses, label='Train Loss')
    plt.plot(steps, val_losses, label='Val Loss')
    plt.xlabel('Steps')
    plt.ylabel('Loss')
    plt.title('Training and Validation Loss')
    plt.legend()
    plt.grid(True)
```
