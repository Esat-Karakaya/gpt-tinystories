```python
from transformers import AutoTokenizer
```

```python
model_name = "abakirci/admbkrc-turkish-tokenizer"

tokenizer = AutoTokenizer.from_pretrained(model_name)
tokenizer.vocab_size
```

```python
def tokenize(tokenizer, str):
    idx = tokenizer.encode(str)
    res = ""
    for id in idx:
        res+=tokenizer.decode(id)+"\\"
    return res, len(idx)
```

```python
story1="Tom ve Mia mutluydular. Anne ve baba ile pizzacıya gittiler. Pizzacı çok güzeldi. İçeride birçok insan vardı.\n\nGarson geldi. \"Ne yemek istersiniz?\" diye sordu. Anne ve baba peynirli pizza istedi. Tom ve Mia ise sucuklu pizza istedi. Pizzalar geldi! Hepsi çok lezzetliydi.\n\nPizza yedikten sonra dışarı çıktılar. Gökyüzü karanlıktı ve yıldızlar parlıyordu. Çok güzeldi! Hepsi yıldızlara baktı ve gülümsedi.\n\nEve döndüler. Hepsi çok mutluydu. \"Pizza çok güzeldi!\" dedi Tom. \"Evet!\" dedi Mia. Anne ve baba da gülümsedi. Hepsi iyi geceler diledi ve yattı.\n"
```

```python
txt, size = tokenize(tokenizer, story1) # type: ignore
print(txt)
print(size)
```
