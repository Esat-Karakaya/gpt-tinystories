```python
import json
from tqdm import tqdm
```

```python
input_file = "/home/esat/Desktop/gpt-tinystories/datasets/TinyStories_Turkish/input.jsonl"
```

```python
stories=[]
with open(input_file, "r") as file:
    for line in tqdm(file):
        obj = json.loads(line)
        try:
            stories.append(obj["response"]["candidates"][0]["content"]["parts"][0]["text"])
        except:
            try:
                print(obj["response"]["promptFeedback"]["blockReason"])
            except:
                print(obj["response"]["candidates"][0]["finishMessage"])
```

```python
from datasets import Dataset
```

```python
dataset = Dataset.from_dict({
    "text": stories
})
```

```python
dataset.save_to_disk("/home/esat/Desktop/gpt-tinystories/datasets/TinyStories_Turkish/ts_train")
```
