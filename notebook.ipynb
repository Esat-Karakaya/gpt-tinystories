```python
str_tr_ds = """Bir zamanlar Lily adında küçük bir kız varmış. Dışarıda çimlerde oynamayı ve güzel çiçekleri görmeyi seviyordu. Bir gün bir yaprağın üzerinde sürünen küçük bir böcek gördü. Çok küçüktü ve kanatları vardı. Lily bunun çok ilginç olduğunu düşündü.O günün ilerleyen saatlerinde Lily'nin annesi, biraz atıştırmalık alması için onu mağazaya götürdü. Lily rafta büyük bir kavanoz kurabiye gördü. Annesine onları alıp alamayacaklarını sordu ama annesi müsait olmadıklarını söyledi. Lily biraz üzüldü ama sonra daha önce gördüğü böceği hatırladı ve yeniden mutlu oldu."""

str_tr_gpt = """Bir gün, A adında küçük bir kız vardı. Odasında bir iğne buldu. İğne sivriydi, onunla oynamak iyi değildi.  A iğneyi annesine götürmek istedi. Annesi düğmeyi dikebilirdi.  A annesine gitti ve dedi ki: “Anne, bir iğne buldum. Düğmem koptu.”  Annesi gülümsedi ve dedi ki: “Evet A, birlikte düğmeni dikelim.”  Anne ve A birlikte düğmeyi dikti. İşleri kolay oldu. İş bitince A annesine teşekkür etti. İkisi de mutlu oldu. """

str_tr_translation = """Bir gün, A adında küçük bir kız odasında bir iğne buldu. İğnenin sivri olduğu için onunla oynamanın zor olduğunu biliyordu. A, annesiyle iğneyi paylaşmak ve gömleğine bir düğme dikmek istedi.A annesine gitti ve "Anne, bu iğneyi buldum. Benimle paylaşabilir misin ve gömleğimi dikebilir misin?" dedi. Annesi gülümsedi ve "Evet A, iğneyi paylaşabilir ve gömleğini tamir edebiliriz." dedi.Birlikte iğneyi paylaştılar ve A'nin gömleğine düğmeyi diktiler. Paylaştıkları ve birbirlerine yardım ettikleri için bu onlar için zor değildi. İşleri bitirdikten sonra A, iğneyi paylaştığı ve gömleğini tamir ettiği için annesine teşekkür etti. İkisi de paylaştıkları ve birlikte çalıştıkları için mutluydular."""

str_eng="""One day, a little girl named Lily found a needle in her room. She knew it was difficult to play with it because it was sharp. Lily wanted to share the needle with her mom, so she could sew a button on her shirt.Lily went to her mom and said, "Mom, I found this needle. Can you share it with me and sew my shirt?" Her mom smiled and said, "Yes, Lily, we can share the needle and fix your shirt."Together, they shared the needle and sewed the button on Lily's shirt. It was not difficult for them because they were sharing and helping each other. After they finished, Lily thanked her mom for sharing the needle and fixing her shirt. They both felt happy because they had shared and worked together."""
```

```python
def print_tokens(tokenizer, str):
    idx=tokenizer.encode(str)
    print(len(idx))
    for id in idx:
        print(tokenizer.decode(id), end="\\")
    print()
```

```python
tokenizer_bayram = AutoTokenizer.from_pretrained("alibayram/tr_tokenizer", use_fast=True)
```

```python
tokenizer_eng = AutoTokenizer.from_pretrained("roneneldan/TinyStories-1M")
```

```python
tokenizer_vngrs = AutoTokenizer.from_pretrained("vngrs-ai/Kumru-2B")
```

```python
tokenizer_ilker = AutoTokenizer.from_pretrained("kaanilker/mini-turkish-tokenizer")
```

```python
tokenizer_bakirci = AutoTokenizer.from_pretrained("abakirci/admbkrc-turkish-tokenizer")
```
